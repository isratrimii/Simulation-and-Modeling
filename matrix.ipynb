```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns

random_vector = np.random.rand(10)
print(random_vector)

# Plot the histogram
plt.hist(random_vector, bins=50,  alpha=0.5, color='red', edgecolor='green')
# Labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram of Uniformly Distributed Random vectors')
plt.grid(True)

# Show the plot
plt.show()



```

```python
# Generate 500 random values from a normal distribution
SNormalValues = np.random.randn(500)#mean=0, standard deviation=1

# Create a histogram
plt.hist(SNormalValues, bins=30, edgecolor='red')
plt.title('Histogram of Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
